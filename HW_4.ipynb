### Уровень 0:


```
from scipy.misc import derivative
import math as m
from sympy import symbols
import matplotlib.pyplot as plt
import numpy as np
```

Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.



```
def f(x):
    return x**5 + 4 * m.sin(2 * x) + m.cos(3 * x + 3)
```

```
print(f'Значение первой производной {derivative(f, 1.0)}')
print(f'Значение второй производной {derivative(f, 1.0, n=2)}')
```

### Уровень 1:


Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```
y = lambda x: m.sin (2 * x + 1) ** 5
xs = np.linspace(-5, 5)
ys = [y(x) for x in xs]

f1 = [derivative(y, x) for x in xs]
f2 = [derivative(y, x, n=2) for x in xs]
```

```
fig, axs = plt.subplots(3)
axs[0].plot(xs, ys) # График функции
axs[1].plot(xs, f1) # График первой производной функции
axs[2].plot(xs, f2) # График второй производной функции
```
