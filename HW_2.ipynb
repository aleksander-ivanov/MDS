### Уровень 0:

 - Создайте матрицу 3 на 3, заполненную числами от 1 до 9
 - Прибавьте к ней матрицу с единицами на диагонале
 - Умножьте полученную матрицу на вектор (1, 2, 3)
 - Посчитайте обратную матрицу к полученной в результате прибавления матрице

````
import numpy as np
import matplotlib.pyplot as plt
````

````
# Создайте матрицу 3 на 3, заполненную числами от 1 до 9
x =  np.arange(1, 10).reshape(3,3)
print(x)
````

````
# Прибавьте к ней матрицу с единицами на диагонале
y = np.eye(3) 
z = x + y
print(z)             
````

````
# Умножьте полученную матрицу на вектор (1, 2, 3)
v = np.array([1,2,3])

res = z.dot(v)
print(res)
````

````
# Посчитайте обратную матрицу к полученной в результате прибавления матрице
np.linalg.inv(z)
````



При помощи матричных операций выполните:
- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x
- Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1.2)



````
A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)
````

````
x = A[:,0]
y = A[:,1]

print(x)
print(y)
````

````
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

````
# Уменьшить объект в два раза
halved_x = x // 2
halved_y = y // 2

plt.plot(halved_x,halved_y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

````
# Повернуть на 130 градусов
from scipy.ndimage.interpolation import rotate

rotated_x = rotate(x, angle=130)
rotated_y = rotate(x, angle=130)

plt.plot(rotated_x,rotated_y)
plt.ylim([-10, 10])
plt.xlim([-10, 10])
````

````
# Отразить объект относительно прямой y=x
x_flipped = np.flip(x, 0)
y_flipped = np.flip(y, 1)

plt.plot(x_flipped,y_flipped)
plt.ylim([0, 3])
plt.xlim([0, 3])
````

````
# Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1.2)
x_rolled = np.roll(x, (200, 300), axis = (0, 1))
y_rolled = np.roll(y, (200, 300), axis = (0, 1))

x_scaled = np.interp(x_rolled, (x_rolled.min(), x_rolled.max()), (0.5, 1.2))
y_scaled = np.interp(y_rolled, (y_rolled.min(), y_rolled.max()), (0.5, 1.2))

plt.plot(x_scaled,y_scaled)
plt.ylim([0, 3])
plt.xlim([0, 3])
````
